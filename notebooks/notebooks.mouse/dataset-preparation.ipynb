# Non-Human TCR set

Get mouse and macaque data. This will be part of the `Non-Human TCR set`.

```python
import pandas as pd
import os
import numpy as np
login = os.getlogin( )

datadir = f"/home/{login}/Git/tcr/data/"
vdjdb_df = pd.read_csv(os.path.join(datadir, 'vdjdb/vdjdb-2021-09-05/vdjdb.txt'), sep='\t')
vdjdb_df
```

```python
vdjdb_df['species'].unique()
```

```python
df = vdjdb_df[vdjdb_df['species']  != 'HomoSapiens']
df = df[df['gene'] == 'TRB']
df.columns
```

```python
df = df[['antigen.epitope', 'cdr3']]
df.dropna()
len(df)
```

```python
df = df.rename(columns={'antigen.epitope': 'peptide', 'cdr3': 'tcrb'})
```

```python
df['sign'] = 1
df['tcra'] = np.nan
```

```python
df.to_csv(datadir+'vdjdb/vdjdb-2021-09-05/mouse-macaco.csv', index=False)
```
